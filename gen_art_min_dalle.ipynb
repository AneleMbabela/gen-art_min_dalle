```python
from PIL import Image
import numpy as np
from min_dalle import MinDalle
import torch
```

```python
model = MinDalle(
    models_root='./pretrained',
    dtype=torch.float32,
    device='cuda',
    is_mega=True,
    is_reusable=True
)
```

```python
prompt = "Nuclear explosion mus "
```

```python
with open("prompt.txt", "w") as f:
  f.write(prompt)
```

```python
images = model.generate_image(
    text=prompt,
    seed=-1,
    grid_size=1,
    is_seamless=False,
    temperature=1,
    top_k=256,
    supercondition_factor=1,
    is_verbose=False
)
```

```python
ext = "png"
filename = "image"
image_path = "{}.{}".format(filename, ext)
images.save(image_path)
```

```python
from PIL import Image
im = Image.open(image_path)
im.show()
```
